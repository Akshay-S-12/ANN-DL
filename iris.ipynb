```python
#encoding output
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
df['Species']=label.fit_transform(df['Species'])
df
```

```python
#drop id column
df.drop('Id',axis=1,inplace=True)   
df
```

```python
#seperate input and output
x=df.iloc[:,:-1].values
print(x)
x.ndim
```

```python
y=df.iloc[:,-1].values
print(y)
y.ndim
```

```python
#training and testing data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)
x_train.shape
```

```python
#Normalization
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
x_train=scaler.fit_transform(x_train)   
x_test=scaler.transform(x_test)
x_train
```

```python
#model creation
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense
ann=Sequential()
#input layer
ann.add(Dense(units=4,activation='relu'))
#hidden layer
ann.add(Dense(units=5,activation='relu'))
#hidden layer
ann.add(Dense(units=4,activation='relu'))
#output layer
ann.add(Dense(units=3,activation='softmax'))
```

```python
#compile
ann.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
x_train.shape
```

```python
ann.fit(x_train,y_train,epochs=100,batch_size=10)
```

```python
ann.evaluate(x_test,y_test)
```

```python
y_new=ann.predict(scaler.transform([[2.3,1,5,3]]))
y_new
```

```python
#argmax
ind=y_new.argmax()
ind
```

```python
pred=label.inverse_transform([ind])
pred
```
