```python
df.drop(['Name'],axis=1,inplace=True)
df.drop(['Surname'],axis=1,inplace=True)
df
```

```python
#seperate input and output
x=df.iloc[:,:-1].values
print(x)
x.ndim
```

```python
y=df.iloc[:,-1].values
print(y)
y.ndim
```

```python
#training and testing data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)
x_train.shape
```

```python
#Normalization
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
#model creation
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense
#initiliazing the ANN
ann=Sequential()  
#adding the input layer 
ann.add(Dense(units=4,activation='relu'))
#adding the hidden layer    
ann.add(Dense(units=10,activation='relu'))
#adding 2nd hidden layer
ann.add(Dense(units=6,activation='relu'))
#adding the output layer
ann.add(Dense(units=1,activation='sigmoid'))
```

```python
#compile the ANN
ann.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
x_train.shape
```

```python
#fitting the ANN to the training set
ann.fit(x_train,y_train,epochs=100,batch_size=12)
```

```python
#evaluating the model
ann.evaluate(x_test,y_test)
```

```python
y_pred=ann.predict(x_test)
```

```python
ann.summary()
```

```python
y_pred=ann.predict(scaler.transform([[35,3,5,4]]))
y_pred
if y_pred>0.5:
    print("The person has lung cancer")
else:   
    print("The person does not have lung cancer")
```
